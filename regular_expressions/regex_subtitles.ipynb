```python
import pandas as pd
path="/content/drive/MyDrive/SMUC_Notebooks /categorized_sentences.xlsx"
data = pd.read_excel(path)
```

```python
most_common_phrases_by_category = data.groupby('Category')['Sentence'].apply(lambda x: x.value_counts().head(5))
most_common_phrases_by_category
```

```python
# To remove all numerical variables (numbers) from the 'Sentence' column, we can use regular expressions with the str.replace method
data['Sentence'] = data['Sentence'].str.replace(r'\d+', '', regex=True)

# Displaying the first few rows of the modified dataframe to verify the changes
data.head()
```

```python
most_common_phrases_by_category = data.groupby('Category')['Sentence'].apply(lambda x: x.value_counts().head(5))
most_common_phrases_by_category
```

```python
# To remove the dash "-" along with any surrounding spaces, we use the str.replace method again
data['Sentence'] = data['Sentence'].str.replace(r'\s*-\s*', ' ', regex=True)

# Displaying the first few rows of the modified dataframe to verify the changes
data.head()
```

```python
most_common_phrases_by_category = data.groupby('Category')['Sentence'].apply(lambda x: x.value_counts().head(5))
most_common_phrases_by_category
```
